# OpenCV Tutorial
## Basic Operations and Image Processing

This notebook covers basic image operations and traditional computer vision techniques using OpenCV.

### 1. Importing OpenCV
First, we need to import the OpenCV library. We'll also import matplotlib to display images in the notebook.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Helper function to display images in the notebook
def show_image(title, img, cmap=None):
    plt.figure(figsize=(6, 4))
    if cmap:
        plt.imshow(img, cmap=cmap)
    else:
        # OpenCV uses BGR, Matplotlib uses RGB
        plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.axis('off')
    plt.show()
```

### 2. Loading an Image
We load an image using `cv2.imread()`. The flag `1` loads a color image, while `0` loads it in grayscale.

```python
# Make sure image.jpg exists in the same directory
img_path = "image.jpg"

# Load as color image
img_color = cv2.imread(img_path, 1)

if img_color is not None:
    show_image("Original Image", img_color)
else:
    print("Image not found. Please ensure image.jpg is in the directory.")
```

### 3. Image Shape / Resolution
The shape of the image gives us the height, width, and number of channels.

```python
if img_color is not None:
    print("Image Shape:", img_color.shape)
    print(f"Height: {img_color.shape[0]} pixels")
    print(f"Width: {img_color.shape[1]} pixels")
    print(f"Channels: {img_color.shape[2]}")
```

### 4. Resizing the Image
We can resize the image using `cv2.resize()`.

```python
if img_color is not None:
    # Resize to 500x300 (width x height)
    resized_img = cv2.resize(img_color, (500, 300))
    show_image("Resized Image", resized_img)
```

### 5. Converting to Grayscale
We can convert a color image to grayscale using `cv2.cvtColor()`.

```python
if img_color is not None:
    gray_img = cv2.cvtColor(img_color, cv2.COLOR_BGR2GRAY)
    show_image("Grayscale Image", gray_img, cmap='gray')
```

### 6. Edge Detection (Canny)
Edge detection is a fundamental image processing technique.

```python
if img_color is not None:
    edges = cv2.Canny(gray_img, 100, 200)
    show_image("Edge Detection", edges, cmap='gray')
```

### 7. Image Histograms
Computing, visualizing, and equalizing histograms for grayscale or RGB images.

```python
if img_color is not None:
    # Compute and visualize histogram for grayscale image
    hist = cv2.calcHist([gray_img], [0], None, [256], [0, 256])
    
    plt.figure(figsize=(6, 4))
    plt.plot(hist)
    plt.title('Grayscale Histogram')
    plt.xlabel('Bins')
    plt.ylabel('# of Pixels')
    plt.show()

    # Equalization
    eq_img = cv2.equalizeHist(gray_img)
    show_image("Equalized Image", eq_img, cmap='gray')
```

### 8. Color Space Conversions
Splitting RGB channels and analyzing pixel intensities.

```python
if img_color is not None:
    # Splitting into B, G, R channels
    b, g, r = cv2.split(img_color)
    
    # Show Red channel in grayscale
    show_image("Red Channel", r, cmap='gray')
    
    # We can also convert to HSV
    hsv_img = cv2.cvtColor(img_color, cv2.COLOR_BGR2HSV)
    show_image("HSV Image", cv2.cvtColor(hsv_img, cv2.COLOR_HSV2RGB))
```

### 9. Sampling, Quantization, and Resolution
Downsampling and altering resolution.

```python
if img_color is not None:
    # Downsampling by a factor of 4
    down_img = cv2.resize(img_color, (img_color.shape[1]//4, img_color.shape[0]//4))
    show_image("Downsampled Image", down_img)
    print("Downsampled shape:", down_img.shape)
```

### 10. Image Noise
Generating and adding synthetic noise to images (e.g., Gaussian noise).

```python
if img_color is not None:
    # Generate Gaussian noise
    row, col, ch = img_color.shape
    mean = 0
    var = 500
    sigma = var**0.5
    gauss = np.random.normal(mean, sigma, (row, col, ch))
    gauss = gauss.reshape(row, col, ch)
    
    # Add noise
    noisy_img = np.clip(img_color + gauss, 0, 255).astype(np.uint8)
    show_image("Gaussian Noise Added", noisy_img)
```

### 11. Image Smoothing and Denoising
Applying blur techniques to remove noise.

```python
if img_color is not None:
    # Smoothing via Gaussian Blur
    blur_img = cv2.GaussianBlur(noisy_img, (5, 5), 0)
    show_image("Gaussian Blurred (Denoised)", blur_img)
    
    # Median Blur (great for salt and pepper noise)
    median_blur = cv2.medianBlur(noisy_img, 5)
    show_image("Median Blurred", median_blur)
```

### 12. Image Filtering & Convolution
Creating custom matrices (kernels) to modify pixels based on local neighborhoods.

```python
if img_color is not None:
    # Custom sharpening kernel
    kernel_sharpening = np.array([[-1,-1,-1], 
                                  [-1, 9,-1],
                                  [-1,-1,-1]])
    
    # Applying the custom kernel via 2D Convolution
    sharpened = cv2.filter2D(img_color, -1, kernel_sharpening)
    show_image("Sharpened Custom Kernel", sharpened)
```

### 13. Edge Detection (Derivatives and Gradients)
Computing image derivatives (Sobel and Laplacian) to identify hard edges.

```python
if img_color is not None:
    # Sobel Edge Detection
    sobelx = cv2.Sobel(gray_img, cv2.CV_64F, 1, 0, ksize=5)  # X-direction
    sobely = cv2.Sobel(gray_img, cv2.CV_64F, 0, 1, ksize=5)  # Y-direction
    
    # Laplacian Edge Detection
    laplacian = cv2.Laplacian(gray_img, cv2.CV_64F)
    
    show_image("Sobel X", cv2.convertScaleAbs(sobelx), cmap='gray')
    show_image("Sobel Y", cv2.convertScaleAbs(sobely), cmap='gray')
    show_image("Laplacian", cv2.convertScaleAbs(laplacian), cmap='gray')
```

### 14. Template Matching
Detecting specific patterns or sub-images within a larger image.

```python
if img_color is not None:
    # Selecting a random patch as a template
    template = gray_img[100:200, 100:200]
    if template.size > 0:
        res = cv2.matchTemplate(gray_img, template, cv2.TM_CCOEFF_NORMED)
        
        # Get optimal match
        min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(res)
        top_left = max_loc
        bottom_right = (top_left[0] + template.shape[1], top_left[1] + template.shape[0])
        
        # Draw rectangle
        matched_img = img_color.copy()
        cv2.rectangle(matched_img, top_left, bottom_right, (0, 255, 0), 2)
        show_image("Template Match", matched_img)
```

### 15. Facial Detection
Identifying the bounding boxes of faces within an image using Haar Cascades.

```python
if img_color is not None:
    # Load pre-trained Haar cascade for face detection
    face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
    
    # Detect faces
    faces = face_cascade.detectMultiScale(gray_img, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))
    
    face_img = img_color.copy()
    for (x, y, w, h) in faces:
        cv2.rectangle(face_img, (x, y), (x+w, y+h), (255, 0, 0), 2)
        
    show_image("Facial Detection", face_img)
```
