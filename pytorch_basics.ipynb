# PyTorch Basics for Beginners

Welcome to the PyTorch Basics tutorial! PyTorch is a popular open-source machine learning and deep learning framework. At its core, PyTorch provides two high-level features:
1. Multi-dimensional arrays (called **Tensors**) similar to NumPy, but with strong acceleration via GPUs.
2. Automatic differentiation (Autograd) for building and training neural networks.

Let's dive into the fundamental building block: **Tensors**.

### 1. Importing PyTorch
First, let's import the PyTorch library. We usually import it as `torch`.

```python
import torch
import numpy as np

print(f"PyTorch Version: {torch.__version__}")
```

### 2. Introduction to Tensors
A tensor is a generalization of vectors and matrices and is easily understood as a multidimensional array.
- A scalar is a zero-dimensional tensor
- A vector is a 1D tensor
- A matrix is a 2D tensor
- An N-dimensional array is an ND tensor

Let's create some basic tensors.

```python
# Creating an uninitialized 2x3 tensor (matrix)
x = torch.empty(2, 3)
print("Uninitialized Tensor:\n", x)

# Creating a tensor with random values between 0 and 1
rand_tensor = torch.rand(2, 3)
print("\nRandom Tensor:\n", rand_tensor)

# Creating a tensor filled with zeros (useful for biases)
zeros_tensor = torch.zeros(2, 3, dtype=torch.float32)
print("\nZeros Tensor:\n", zeros_tensor)
```

### 3. Creating Tensors from Data
You can create a tensor directly from a standard Python list or a NumPy array.

```python
# From a Python List
my_list = [[1, 2, 3], [4, 5, 6]]
tensor_from_list = torch.tensor(my_list)
print("Tensor from List:\n", tensor_from_list)

# From a NumPy Array
np_array = np.array([7, 8, 9])
tensor_from_np = torch.from_numpy(np_array)
print("\nTensor from NumPy:\n", tensor_from_np)
```

### 4. Tensor Properties (Shape, Type, Device)
Every tensor has three very important attributes:
- `shape`: The dimensions of the tensor.
- `dtype`: The data type (e.g., float32, int64).
- `device`: Where the tensor is stored (CPU or GPU/CUDA).

```python
t = torch.rand(3, 4)

print("Shape of tensor:", t.shape)
print("Datatype of tensor:", t.dtype)
print("Device tensor is stored on:", t.device)
```

### 5. Basic Math Operations
You can perform standard mathematical operations on tensors. PyTorch supports standard operators like `+`, `-`, `*`, `/`.

```python
a = torch.tensor([1, 2, 3])
b = torch.tensor([4, 5, 6])

# Addition
print("Addition:", a + b)

# Subtraction
print("Subtraction:", a - b)

# Multiplication (Element-wise)
print("Element-wise Multiplication:", a * b)

# Matrix Multiplication (dot product for 1D tensors)
dot_product = torch.dot(a, b)
print("Dot Product:", dot_product)
```

### 6. Reshaping Tensors (View)
Often, you need to change the shape of your tensor (e.g., flattening an image before passing it into a linear layer). PyTorch uses `.view()` or `.reshape()`.

```python
x = torch.rand(4, 4)
print("Original Shape:", x.shape)

# Flatten the tensor to 1D
y = x.view(16)
print("Flattened Shape:", y.shape)

# Reshape to 2x8. Using -1 tells PyTorch to automatically calculate that dimension
z = x.view(-1, 8)
print("Reshaped Shape (-1, 8):", z.shape)
```

### 7. Moving Tensors to the GPU (CUDA)
By default, tensors are created on the CPU. If you have a GPU, you can move them to speed up mathematical computations.

```python
# Check if CUDA (GPU) is available
if torch.cuda.is_available():
    device = torch.device("cuda")
    print("GPU is available. Moving tensor to GPU...")
    
    # Create tensor directly on the device
    x_gpu = torch.ones(3, 3, device=device)
    
    # Or move an existing tensor
    y_cpu = torch.zeros(3, 3)
    y_gpu = y_cpu.to(device)
    
    print("x_gpu device:", x_gpu.device)
else:
    print("GPU not available. Continuing with CPU.")
```

### 8. Autograd (Automatic Differentiation)
PyTorch has a built-in tracking engine called `torch.autograd`. It automatically computes gradients (derivatives) which is essential for training neural networks. 
Set `requires_grad=True` to tell PyTorch to track operations on a tensor.

```python
# Create a tensor with requires_grad=True
x = torch.tensor([2.0], requires_grad=True)

# Run a computation: y = x^2 + 3x + 1
y = x**2 + 3*x + 1
print("y:", y)

# Compute the gradient (dy/dx)
y.backward()

# The derivative of x^2 + 3x + 1 is 2x + 3. 
# So if x = 2, dy/dx = 2(2) + 3 = 7.
print("Gradient (dy/dx) at x=2:", x.grad)
```
